## Analysis Starts for Insights

### 1. Below statement shows the top 10 products according to ratings

```python
amazon.sort_values('rating',ascending=False).head(10)
```

### 2. Below statement is for products with high rating as well as higher rating_counts(more than 2000)

```python
amazon[amazon['rating_count'] >= 2000].sort_values(['rating', 'rating_count'], ascending=[False, False]).head(10)
```

### 3. Below shows data with high rating and discount

```python
high_value_products = amazon[(amazon['rating'] >= 4.7) & (amazon['discount_percentage'] >= 0.60)]
```

```python
high_rating_with_high_discount = high_value_products.sort_values(['rating', 'discount_percentage'])
high_rating_with_high_discount
```

### 4. Average discount for each category(category_1)

```python
avg_discount_category1=amazon.groupby('category_1')['discount_percentage'].agg(['min','max','mean'])
avg_discount_category1
```

### 5. Average discount for each category(category_2)

```python
avg_discount_category2=amazon.groupby('category_2')['discount_percentage'].agg(['min','max','mean'])
avg_discount_category2
```

### 6. Most popular product category wise(category_1)

```python
most_popular_product=amazon.groupby('category_1')['rating_count'].agg('sum').sort_values(ascending=False)
most_popular_product
```

### 7. Most popular product category wise(category_3) based on rating_count

```python
most_popular_product=amazon.groupby('category_3')['rating_count'].agg('sum').sort_values(ascending=False)
most_popular_product
```

### 8. Most popular product category wise(category_3) based on rating

```python
most_popular_product=amazon.groupby('category_3')['rating'].agg('sum').sort_values(ascending=False)
most_popular_product
```

### 9. Comparing for high actual_price with high ratings

```python
high_actual_value_products = amazon[(amazon['rating'] >= 4.7) & (amazon['actual_price'] >= 1000)]
```

```python
high_rating_with_high_actual_price = high_actual_value_products.sort_values(['rating', 'actual_price'])
high_rating_with_high_actual_price
```

### 10. Average price for products with high rating as well as high rating_count

```python
highest_rated_products=amazon.groupby(['rating','rating_count'])['actual_price'].agg('mean').sort_values(ascending=False)
highest_rated_products.sort_values()
```

### 11. Top 10 most popular products

```python
top10_popular=amazon.sort_values('rating_count',ascending=False).head(10)
top10_popular
```

### 12. Most Selling category based on actual_price

```python
most_selling_category=amazon.groupby('category_1')[['actual_price','rating_count']].agg('sum')
most_selling_category
```

### 13. Most Selling sub-category

```python
most_selling_sub_category=amazon.groupby('category_2')[['actual_price','rating_count']].agg('sum')
most_selling_sub_category
```

### 14. Product with high discount and high sales(based on ratings_count)

```python
high_discount_products = amazon[(amazon['discount_percentage'] >= 0.70) & (amazon['rating_count'] >= 10000)]
high_rating_with_high_discount = high_discount_products.sort_values(['discount_percentage', 'rating_count'])
high_rating_with_high_discount
```

### 15. Rating Category wise

```python
rating_category=amazon.groupby('category_1')['rating'].mean().sort_values(ascending=False)
rating_category
```

### 16. Rating sub-category wise 

```python
rating_sub_category=amazon.groupby('category_2')['rating'].mean().sort_values(ascending=False)
rating_sub_category
```

### 17. Checking for products with high price as well as high ratings

```python
high_actual_price_products = amazon[(amazon['actual_price'] >= 10000) & (amazon['rating_count'] >= 10000)]
high_rating_with_high_actual_price = high_actual_price_products.sort_values(['actual_price', 'rating_count'])
high_rating_with_high_actual_price
```

### 18. High Rating High Rating count but low actual price

```python
low_actual_price_products = amazon[(amazon['actual_price'] <= 10000) & (amazon['rating_count'] >= 10000) & (amazon['rating'] >= 4.5)]
high_rating_with_low_actual_price = low_actual_price_products.sort_values(['actual_price', 'rating_count','rating'])
high_rating_with_low_actual_price
```

### 19. High actual price but low ratings

```python
high_actual_price_products = amazon[(amazon['actual_price'] >= 10000) & (amazon['rating'] <= 3.5)]
low_rating_with_high_actual_price = high_actual_price_products.sort_values(['actual_price', 'rating'])
low_rating_with_high_actual_price
```

### 20. Discount category wise with high rating_count

```python
category = amazon.groupby('category_1').agg(avg_discount=('discount_percentage', 'mean'),total_rating_count=('rating_count', 'sum'))
category_wise_with_high_rating = category.sort_values('total_rating_count',ascending=False)
category_wise_with_high_rating
```
